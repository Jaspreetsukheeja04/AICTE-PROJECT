```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import string

# Take first 5000 characters for summarization
short_text = pdf_text[:5000]

def summarize_text(text):
    sentences = sent_tokenize(text)

    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text.lower())
    filtered_words = [w for w in words if w.isalpha() and w not in stop_words and w not in string.punctuation]

    word_counts = Counter(filtered_words)
    top_words = [word for word, count in word_counts.most_common(20)]

    # Pick sentences containing top words
    summary_sentences = [sent for sent in sentences if any(word in sent.lower() for word in top_words)]

    # Return first 10 sentences as summary
    return " ".join(summary_sentences[:10])

# Generate summary
summary = summarize_text(short_text)

print("===== SUMMARY =====\n")
print(summary)

```

```python
def generate_mcqs(text, num_questions=5):

    sentences = sent_tokenize(text)
    questions = sentences[:num_questions]

    mcqs = []

    for i, sentence in enumerate(questions):
        words = sentence.split()
        if len(words) > 6:
            answer = words[-1].strip(".")
            question = sentence.replace(answer, "_____")

            options = [answer, "Protein", "DNA", "Enzyme"]
            mcqs.append((question, options, answer))

    return mcqs
```

```python
mcqs = generate_mcqs(summary)

print("\n===== MCQs =====\n")

for i, (q, options, ans) in enumerate(mcqs):
    print(f"Q{i+1}: {q}")
    for opt in options:
        print("-", opt)
    print()
```

```python
# Clean Quiz function
import string

def start_quiz(mcqs):
    score = 0

    for i, (q, options, ans) in enumerate(mcqs):
        # Remove extra symbols from question
        clean_q = q.translate(str.maketrans('', '', string.punctuation.replace('_', '')))
        print(f"\nQ{i+1}: {clean_q}")

        # Clean options
        clean_options = [opt.translate(str.maketrans('', '', string.punctuation)) for opt in options]

        # Print options as A/B/C/D
        for idx, opt in enumerate(clean_options):
            print(f"{chr(65+idx)}. {opt}")

        # User input
        user = input("Your answer (A/B/C/D): ").strip().upper()

        # Map user input to option text
        try:
            user_text = clean_options[ord(user)-65]
        except:
            user_text = ""

        # Check answer
        if user_text.lower() == ans.lower():
            print("Correct!")
            score += 1
        else:
            print("Wrong! Correct answer was:", ans)

    print(f"\nFinal Score: {score} / {len(mcqs)}")


# Call the quiz
start_quiz(mcqs)
```
